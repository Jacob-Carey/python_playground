```python
from sklearn import tree
```

```python
from sklearn import preprocessing
```

```python
import pandas as pd
```

```python
import pydotplus as pypl
```

```python
from IPython.display import Image
```

```python
le = preprocessing.LabelEncoder()
```

```python
df = pd.read_csv("C:/Users/jacob.carey/Desktop/ConsentDataOutput_100.csv", header=0, encoding='utf-8', index_col=0)
```

```python
features = [
            'ConsentDurationMonths', 
            'ConsentRequired', 
            'ConsentType', 
            'Country', 
            'LanguageStyle', 
            'NavigateConsentEnabled', 
            'OnBrand', 
            'PageClickConsentEnabled', 
            'ScrollConsentEnabled'
            ]
```

```python
clf = tree.DecisionTreeClassifier(min_samples_split=100)
```

```python
clf = clf.fit(df[features], df['Consent_bool'])
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=features, filled=True, 
                                rounded=True, class_names=['nonconsent', 'consent'],
                                proportion=True, node_ids=True, max_depth=3)
```

```python
graph = pypl.graph_from_dot_data(dot_data)
```

```python
graph.write_png("dtree.png")
```
