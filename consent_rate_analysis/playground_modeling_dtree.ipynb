```python
from sklearn import tree
```

```python
import pandas as pd
```

```python
import pydotplus as pypl
```

```python
from IPython.display import Image
```

```python
from datetime import datetime
```

```python
df = pd.read_csv("C:/Users/jacob.carey/Desktop/ConsentDataOutput_1.csv", header=0, encoding='utf-8', index_col=0)
```

```python
features = [
            #'ConsentDurationMonths', 
            #'ConsentRequired', 
            'ConsentType', 
            'Country', 
            'LanguageStyle', 
            #'NavigateConsentEnabled', 
            'OnBrand', 
            #'PageClickConsentEnabled', 
            #'ScrollConsentEnabled'
            ]
```

```python
clf = tree.DecisionTreeClassifier(max_depth=5)
```

```python
clf = clf.fit(df[features], df['Consent_bool'])
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=features, filled=True, 
                                rounded=True, class_names=['nonconsent', 'consent'],
                                proportion=True, node_ids=True)
```

```python
graph = pypl.graph_from_dot_data(dot_data)
```

```python
date_time = datetime.today().strftime('%m-%d-%H%M')
```

```python
graph.write_png("consent_rate_dtree_{0}.png".format(date_time))
```
