```python
import requests
import json
import pprint
```

```python
uid = '7ff1fd51-8431-420c-9319-ff85247a087f'
scan_region_id = '11'
scanurl = 'https://www.google.com'
```

```python
data = {
        'ClientIdentifier': 'null',
        'CountryId': '1',
        'ProxyUrl': 'null',
        'ScanRegionId': scan_region_id,
        'TestDrive': 'false',
        'Url': scanurl,
        'UserId': uid
    }
```

```python
scan_id_request = requests.post('https://trackermapapi.evidon.com/api/TrackerMap/ScanUrl', data)
scan_response = scan_id_request.json()
scan_id = scan_response['UniqueScanId']
```

```python
params = {
    'scanRegionId': scan_region_id,
    'scanId': scan_id
}
```

```python
node_map_request = requests.post('https://trackermapapi.evidon.com/api/TrackerMap/GetNodeMap?', params=params)
node_response = node_map_request.json()
```
