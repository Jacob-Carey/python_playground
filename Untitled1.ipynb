```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/jacob.carey/Documents/consentv2/ba_pull.csv", header=0, usecols={
"Id",
"company_name",
"Domain",
"SiteNoticeThemeId",
"Industry",
"ConsentToolThemeId",
"ConsentDisplayTypeId",
"IncludeSubdomains",
"ConsentTypeId",
"PrivacyAccessTypeId",
"ConsentDurationMonths",
"L2Enabled",
"GDPREnabled",
"IABEnabled",
"ConsentDetailCategories",
"ConsentDetailVendors",
"AdChoicesEnabled",
"CountryId",
"country_name",
"Identifier",
#"TranslatedText"
})
```

```python
df['Screenshot'] = ''
```

```python
newlist = []
```

```python
for row in df.iterrows():
    my_dict = dict(row[1])
    country = my_dict['country_name']
    
    if country == 'United States':
        country = 'United%States'
   
    if country == 'United Kingdom':
        country = 'United%Kingdom'
        
    if country == 'Czech Republic':
        country = 'Czech%Republic'
        
    id = my_dict['Id']
    url = "https://crownpeakgroup.sharepoint.com/sites/ConsentRateAnalysisProject/Shared%20Documents/screenshotsv1/{0}/{1}.png".format(country, id)
    
    my_dict['Screenshot'] = url
    
    newlist.append(my_dict)
```

```python
new_df = pd.DataFrame(newlist)
```
