```python
import pandas as pd
```

```python
my_df = pd.read_csv("C:/Users/jacob.carey/Desktop/DataSets/SiteNotice/individual_user_eu_sample_pull.csv", header=0,
                             usecols=[
                                 "consent_time_utc",
                                 "protocol",
                                 "host_domain",
                                 "path",
                                 "country_code",
                                 "country",
                                 "region",
                                 "city",
                                 "user_ip",
                                 "company_id",
                                 "notice_id",
                                 "country_id",
                                 "os",
                                 "browser",
                                 "device",
                                 "consent_given",
                                 "consent_request_type",
                                 "consent_action",
                                 "filedate"
                             ])
```

```python
newlist = []
sample_ratio = 100
```

```python
my_df.head(50)
```

```python
for row in my_df.iterrows():
    my_dict = dict(row[1])
    
    #Cleaning up data types and data values
    
    #format Country to numeric value
    test_val = my_dict.get('Country', 0).lower().strip()
    if test_val == 'uk': 
        my_dict['Country'] = 1
    elif test_val == 'united kingdom': 
        my_dict['Country'] = 1
    elif test_val == 'france': 
        my_dict['Country'] = 2
    elif test_val == 'germany': 
        my_dict['Country'] = 3
    
    #format OnBrand to numeric value 
    test_val = my_dict.get('OnBrand', 0).lower().strip()
    if test_val == 'no': 
        my_dict['OnBrand'] = 1
    elif test_val == 'yes': 
        my_dict['OnBrand'] = 2
    elif test_val == 'partial': 
        my_dict['OnBrand'] = 3

    
    #format LanguageStyle to numeric value
    test_val = my_dict.get('LanguageStyle', 0).lower().strip()
    if test_val == '?': 
        my_dict['LanguageStyle'] = 99
    elif test_val == 'unclear': 
        my_dict['LanguageStyle'] = 0
    elif test_val == 'clear': 
        my_dict['LanguageStyle'] = 1

    
    #copies my_dict, removes Users and Consent, 
    #fills newlist with consenting and nonconsenting samples / sample_ratio
    #creates dataframe new_df with user samples 
    c_dict = my_dict.copy()
    del c_dict['Users']
    del c_dict['Consent']
    c_dict['Consent_bool'] = 1
    for idx in range(int(my_dict['Consent']/sample_ratio)):     
        newlist.append(c_dict)
    nc_dict = c_dict.copy()
    nc_dict['Consent_bool'] = 0
    nc_count = my_dict['Users'] - my_dict['Consent']
    for idx in range(int(nc_count/sample_ratio)):
        newlist.append(nc_dict)
```

```python
new_df = pd.DataFrame(newlist)
```

```python
new_df.to_csv(path_or_buf="C:/Users/jacob.carey/Desktop/SiteNotice Consent Rate Analysis/ConsentDataOutput_{0}.csv".format(sample_ratio))
```
