```python
import urllib 
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Global_Industry_Classification_Standard'
```

```python
data = []
```

```python
# for pg in url:
#     r = urllib.request.urlopen(pg)
```

```python
r = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(r, 'html.parser')
```

```python
first = soup.find_all('td')
```

```python
first
```

```python
for td in first:
    x = td.get_text()
    data.append(x)
```
